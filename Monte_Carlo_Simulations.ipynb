**Predicting the firm's future gross profit**



```
import numpy as np
import matplotlib.pyplot as plt
```

```
#rev mean
rev_m=170
rev_std=20
iterations=1000
```

```
rev=np.random.normal(rev_m,rev_std,iterations)
rev
```

```
plt.figure(figsize=(15,10))
plt.plot(rev)
```

```
COGS=-(rev*np.random.normal(0.6,0.1))
plt.figure(figsize=(15,6))
plt.plot(COGS)
```

```
COGS.mean()
```

```
COGS.std()
```

```
Gross_Profit=rev+COGS
Gross_Profit
plt.figure(figsize=(15,10))
plt.plot(Gross_Profit)
plt.show()
```

```
max(Gross_Profit)
```

```
min(Gross_Profit)
```

```
plt.figure(figsize=(10,6))
plt.hist(Gross_Profit,bins=[40,50,60,70,80,90,110,120])
```
